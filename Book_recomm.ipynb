```python
import pandas as pd
import numpy as np

Book = pd.read_csv("Books.csv")
Book.info()
Book.head()
```

```python
Rating  =  pd.read_csv("Ratings.csv")
Rating.info()
Rating.head()
```

```python
User = pd.read_csv("Users.csv")
User.info()
User.head()
```

```python
Book_list= Book.merge(Rating,on="ISBN")
```

```python
x = Book_list.groupby("Book-Title")['Book-Rating'].count().reset_index()
```

```python

x = x[x["Book-Rating"]>250].sort_values("Book-Rating",ascending=False)
```

```python
avgerage_rating= Book_list.groupby("Book-Title")['Book-Rating'].mean().reset_index()
```

```python
avgerage_rating = avgerage_rating[avgerage_rating['Book-Title'].isin(x['Book-Title'].values)]
```

```python
popularity= x.merge(avgerage_rating,on='Book-Title')
```

```python
popularity.head()
```

```python
popularity.rename(columns={'Book-Rating_x':'Number_of_Ratings','Book-Rating_y':'Avg-Rating'},inplace=True)
```

```python
popularity = popularity.merge(Book,on='Book-Title').drop_duplicates('Book-Title')[['ISBN','Book-Title','Avg-Rating','Book-Author','Year-Of-Publication','Publisher',"Image-URL-M"]]
```

Colaborative recommender system

```python
padhe_likhe_users = Book_list.groupby('User-ID').count()['Book-Title']>=200
```

```python
Book_list.head(3)
```

```python
padhe_likhe_users  = padhe_likhe_users[padhe_likhe_users].index
```

```python
filtered = Book_list[Book_list['User-ID'].isin(padhe_likhe_users)]
```

```python
fi = filtered.groupby('Book-Title')['User-ID'].count()>=50
# filtered_name = fi[fi].index
Book_names = fi[fi].index
```

```python
final_ans = filtered[filtered['Book-Title'].isin(Book_names)]
```

```python
final_ans.drop_duplicates()
```

```python
pt = final_ans.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating').fillna(0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(pt)
```

```python
def recommender(book_name):
    try:
        index = np.where(pt.index==book_name)[0][0]
        lis = sorted(list((enumerate(similarities[index]))),key = lambda x:x[1],reverse=True)[1:3]

        for i in lis:
            print(pt.index[i[0]])
    except:
        print("This book has not in the list,\nPlease try some other movie")
```

```python
recommender("Angela's Ashes (MMP) : A Memoir")
```
